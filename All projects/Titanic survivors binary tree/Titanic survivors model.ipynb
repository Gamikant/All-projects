```python
def compute_entropy(y):
    entropy = 0.
    p_1 = 0
    n = len(y)
    if len(y)==0:
        return 0
    else:
        for i in range(n):
            p_1 = p_1 + y[i]
        p_1 = p_1/n
        if p_1==0 or p_1==1:
            entropy = 0
        else:
            entropy = -(p_1)*math.log(p_1,2)-(1-p_1)*math.log(1-p_1,2)           
    return entropy

def split_dataset(X, node_indices, feature):
    left_indices = []
    right_indices = []
    for i in node_indices:
        if X[i, feature]==1:
            left_indices.append(i)
        elif X[i, feature]==0:
            right_indices.append(i)
    return left_indices, right_indices

def compute_information_gain(X, y, node_indices, feature):  
    # Split dataset
    left_indices, right_indices = split_dataset(X, node_indices, feature)
    X_node, y_node = X[node_indices], y[node_indices]
    X_left, y_left = X[left_indices], y[left_indices]
    X_right, y_right = X[right_indices], y[right_indices]
    information_gain = 0
    information_gain = compute_entropy(y_node)-(len(y_left)/len(y_node)*compute_entropy(y_left)+len(y_right)/len(y_node)*compute_entropy(y_right)) 
    return information_gain

def get_best_split(X, y, node_indices):   
    # Some useful variables
    num_features = X.shape[1]
    # You need to return the following variables correctly
    best_feature = -1
    IG_1 = []
    if compute_entropy(y)==0:
        best_feature=-1
    else: 
        for i in range(num_features):
            IG_1.append(compute_information_gain(X,y,node_indices,i))
        best_feature = np.argmax(IG_1)
    return best_feature

def split(feature):
    df = pd.read_csv(r'C:\Users\prana\Downloads\train.csv')
    df = df.drop(['PassengerId','Name', 'Ticket', 'Cabin'], axis=1)

# Get one hot encoding of columns B
    one_hot = pd.get_dummies(df['Pclass'])
# Drop column B as it is now encoded
    df = df.drop(['Pclass'],axis = 1)
# Join the encoded df
    df = df.join(one_hot)
    one_hot2 = pd.get_dummies(df['Embarked'])
    df = df.drop('Embarked',axis=1)
    df = df.join(one_hot2)
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
    df = df.dropna().astype(int)
    df['Parch'].mask(df['Parch'] >0 ,1, inplace=True)
    df['SibSp'].mask(df['SibSp'] >0 ,1, inplace=True)
    Age = []
    Max = np.zeros((952,2))
    node_indices = list(range(714))
    for i in range(0,81,5):
        df['Age'].mask(df['Age'] >=int(i) ,0, inplace=True)
        df['Age'].mask(df['Age'] >0 ,1, inplace=True)
        for j in range(0,551, 10):
            df['Fare'].mask(df['Fare'] >=int(j) ,0, inplace=True)
            df['Fare'].mask(df['Fare'] >0 ,1, inplace=True)
            X_train = df.drop(['Survived'], axis=1).to_numpy
            y_train = df['Survived'].to_numpy()
            Age.append(compute_information_gain(X_train, y_train, node_indices, feature))
            print(Age[int(56*(i/5)+j/10)])
            Max[int(56*(i/5)+j/10)] = (i,j)
    k = Age.index(max(Age))
    return Max[k]
split(1)
```
